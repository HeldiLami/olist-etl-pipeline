```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../output/olist_clean.csv")

print(df.shape)
df.head()

df["order_status"].value_counts()

df["price_tier"].value_counts().plot(kind="bar", color="purple")
plt.title("Product distribution by price tier")
plt.xlabel("Price tier")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("../output/price_distribution.png")
plt.show()

(df.groupby("order_status")["is_late"].mean() * 100).round(1)
df.groupby("price_tier")["total_revenue"].mean().round(2)
```
